```python
def make_sure_emb():
    data = {
        'ids': [],
        'images': [],
        'embeddings': []
    }

    SIZE = 10

    dataset = torch.load("./data/ModelStealingPub.pt")
    
    shuffling_order = list(range(len(dataset.ids)))
    random_subset = random.sample(shuffling_order, SIZE)


    for i in random_subset:
        image = dataset.imgs[i]
        id = dataset.ids[i]
        
        image.save(f"./data/sure/images/{id}.png")
        result = model_stealing(f"./data/sure/images/{id}.png")

        data["images"].append(image)
        data["ids"].append(id)
        data["embeddings"].append(result)

    file_path = "./data/sure/sure.pt"
    torch.save(data, file_path)
```

```python
import torch
import torchvision.models as models

# Define the path to the checkpoint file
checkpoint_path = './simsiam/checkpoint_0099.pth.tar'

# Load the pretrained ResNet-50 model
model = models.resnet50(pretrained=False)

# Load the checkpoint
checkpoint = torch.load(checkpoint_path, map_location="cpu")

# Load the pretrained weights into the model
model.load_state_dict(checkpoint['state_dict'], strict=False)

num_ftrs = model.fc.in_features
model.fc = torch.nn.Linear(num_ftrs, 512)


# Set the model to evaluation mode
model.eval()

# Optional: Print model architecture
print(model)
```

```python
input_tensor = torch.randn(1, 3, 32, 32)  # Batch size of 1

# Pass the tensor through the model
output = model(input_tensor)

# The output would be the model's prediction
print(output.shape)
```
